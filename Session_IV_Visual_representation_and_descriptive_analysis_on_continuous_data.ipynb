# Frequency chart for continous data

```python
plt.hist(age) # Here we used hist function to produce a frequency chart, the top arry is frequency, the bottom arry is bondary of the invervals.
```

```python
#Frequency chart with defined the number of bars
fig=plt.figure(figsize=(24,6))
plt.hist(age, bins=20, rwidth=0.90) # bins number tell you how many bars are splited to, 'rwidth' give 0.90 of the entire range
plt.xlabel('Age intervals',fontsize=20)
plt.ylabel('Frequency',fontsize=20)
plt.xticks(np.arange(0,150,5),fontsize=12) #here define the intervals of label on the x-axis 0 is lower bound, 150 is the upper bund and 5 is the interval
plt.yticks(np.arange(0,20,2),fontsize=12)
```

```python
#Frequency chart with defined the number of bars
fig=plt.figure(figsize=(24,6))
plt.hist(age, bins=[5,10,18,30,40,50,60,70,80,100,150], rwidth=0.99) # bins number tell you how many bars are splited to, 'rwidth' give 0.90 of the entire range
plt.xlabel('Age intervals',fontsize=20)
plt.ylabel('Frequency',fontsize=20)
plt.xticks(np.arange(0,150,5),fontsize=12) #here define the intervals of label on the x-axis 0 is lower bound, 150 is the upper bund and 5 is the interval
plt.yticks(np.arange(0,20,2),fontsize=12)
```

# Frequency density chart for continous data

```python
# importing necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
 
# plotting histogram for using distplot()
fig=plt.figure(figsize=(24,6))
sns.distplot(a=age, bins=20,kde=False) #Kde=Kernel density estimate, used for probability density estimate
plt.xlabel('Age intervals',fontsize=20)
plt.ylabel('Frequency',fontsize=20)
plt.xticks(np.arange(0,150,5),fontsize=12) #here define the intervals of label on the x-axis 0 is lower bound, 150 is the upper bund and 5 is the interval
plt.yticks(np.arange(0,20,2),fontsize=12)
# visualizing plot using matplotlib.pyplot library
plt.show()
```

# Histogram (frequency density chart)

```python
# importing necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
 
# plotting histogram for using distplot()
fig=plt.figure(figsize=(24,6))
sns.distplot(a=age, bins=20,kde=True) #Kde=Kernel density estimate, used for probability density estimate
plt.xlabel('Age intervals',fontsize=20)
plt.ylabel('Frequency density',fontsize=20)
plt.xticks(np.arange(0,150,5),fontsize=12) #here define the intervals of label on the x-axis 0 is lower bound, 150 is the upper bund and 5 is the interval
plt.yticks(np.arange(0,0.1,0.1),fontsize=12)
# visualizing plot using matplotlib.pyplot library
plt.show()
```

# Cumulative frequency chart

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
age_interval=np.array([0,10,20,30,40,50,60,70,140])
res = stats.cumfreq(age,numbins=age_interval,defaultreallimits=(0, 140)) #numbins=number of up-to, (0,140) gives the range for x=axis
cumfre=np.append([0],res.cumcount)
cumfre
```

```python
age_interval.min()
```

```python
fig = plt.figure(figsize=(24, 10))
age1 = fig.add_subplot(1, 2, 1)
age2 = fig.add_subplot(1, 2, 2)
age1.hist(age, bins=14)
age1.set_title('Histogram')
age2.plot(age_interval, cumfre)
age2.set_title('Cumulative frequency')
age2.set_xlim([age_interval.min(), age_interval.max()])
age2.set_ylim([cumfre.min(), cumfre.max()]+10)
```
